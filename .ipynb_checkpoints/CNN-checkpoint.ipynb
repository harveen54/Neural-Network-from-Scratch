```python


def Convolution(img,filters, filter_size, stride, pad=False):
    
    if(pad):
        padded_image = np.pad(img[:,:,2], (1,1), mode='constant')
    else:
        padded_image = img
    
    horizontal_scans = padded_image.shape[1] - 2
    vertical_scans = padded_image.shape[0] - 2
    
    images_array = []
    
    for k in filters:
        new_image_1 = np.zeros([img.shape[0], img.shape[1],1])
        new_image_2 = np.zeros([img.shape[0], img.shape[1],1])
        new_image_3 = np.zeros([img.shape[0], img.shape[1],1])
        #print(new_image_1.shape)
        
        for i in range(vertical_scans):
            for j in range(horizontal_scans):
                temp_array_0 =  padded_image[i:i+filter_size, j:j+filter_size,0]
                temp_array_1 =  padded_image[i:i+filter_size, j:j+filter_size,1]
                temp_array_2 =  padded_image[i:i+filter_size, j:j+filter_size,2]

                new_image_1[i,j] = np.sum (np.multiply(k, temp_array_0))
                new_image_2[i,j] = np.sum (np.multiply(k, temp_array_1))
                new_image_3[i,j] = np.sum (np.multiply(k, temp_array_2))
        
        images_array.append(new_image_1)
        images_array.append(new_image_2)
        images_array.append(new_image_3)
    return images_array


arr = Convolution(img, filters, 3,1)
```
